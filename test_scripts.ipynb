```python
import networkx as nx
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
```

```python
edge_df = pd.read_csv("opsahl-powergrid/edge_list.csv", delimiter=",")
node_df = pd.read_csv("opsahl-powergrid/node_list.csv", delimiter=",")
G = nx.from_pandas_edgelist(edge_df, source="Source", target="Target", create_using=nx.Graph())
```

```python
from scripts.summary import node_summary

centrality = node_summary(G)
```

```python
centralities = [c for c in centrality.values()] 
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))

# Plotting without log-scale 
ax1.hist(centralities, bins='sqrt')
ax1.set(xlabel='Betweeness Centrality', ylabel='Betweeness Centrality Distribution', title='y without log scale')

# Plotting with log-scale
ax2.hist(centralities, color = (0.572549,0.2862,0.0,1), bins='sqrt')
# set yscale; can also use plt.yscale('log') or plt.yscale('symlog')
ax2.set(yscale='log', xlabel='Betweeness Centrality', ylabel='Betweeness Centrality Distribution', title='y with log scale')

fig.tight_layout()
plt.show()
```
